```python
import numpy as np 
import pandas as pd 
import torch
from transformers import AutoTokenizer, AutoModel 
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased",)
```

```python
model = AutoModel.from_pretrained("bert-base-uncased",output_hidden_states=True)
```

```python
def get_vector(text,token_length):
    tokens=tokenizer(text,max_length=token_length,padding='max_length',truncation=True)
    output=model(torch.tensor(tokens.input_ids).unsqueeze(0),
                 attention_mask=torch.tensor(tokens.attention_mask).unsqueeze(0)).hidden_states[-1]
    return torch.mean(output,axis=1).detach().numpy()
```

```python
def similarity(text1,text2,token_length=10):
  text1_out = get_vector(text1,token_length=token_length)
  text2_out = get_vector(text2,token_length=token_length)
  return cosine_similarity(text1_out,text2_out)[0][0]
```

```python
one = "rainy days are beautiful"
two = "a beautiful day is the one when it rains"
```

```python
similarity(one,two)
```
